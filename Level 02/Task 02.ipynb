```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
rest_data = pd.read_csv(filepath_or_buffer=r"C:\Users\91882\Desktop\Cognifyz\Dataset .csv")
```

# Level 02

## Task 02 : Cuisine Combination

### 1. Identify the most common combinations of cuisines in the dataset .

```python
comb_cuisines = rest_data["Cuisines"].value_counts().sort_values(ascending=False).head(5)
print(f"the most common combinations of cuisines in the dataset are : {comb_cuisines}")
```

### Bar Chart :

```python
plt.figure(figsize=(7,4))
plt.ylim(0,1000)
plt.bar(comb_cuisines.index , comb_cuisines.values)
plt.xlabel("Cuisines")
plt.ylabel("Counts")
plt.title("common combinations of cuisines")
plt.xticks(rotation = 90 ,ha = 'right')
plt.show()
```

### 2. Determine if certain cuisine combinations tend to have higher ratings :

```python
rated_data = rest_data[rest_data["Aggregate rating"] > 0]
avg_rating_cuisine = rated_data.groupby("Cuisines")["Aggregate rating"].mean().sort_values(ascending = False)
avg_rating_cuisine
```

```python
top_cuisine_avg =round( avg_rating_cuisine.loc[comb_cuisines.index],2).sort_values(ascending = False)
print(f"certain cuisine combinations tend to have higher ratings are : {top_cuisine_avg}")
```
