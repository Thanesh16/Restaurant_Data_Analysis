## Task 01 : Restaurant Ratings

### 1. Analyze the distribution of aggregate ratings and determine the most common rating range .

```python
agg_ratings = rest_data["Aggregate rating"].value_counts()
agg_ratings
common_rating = agg_ratings.idxmax()
common_rating_count = agg_ratings.max()
print(f"The most common rating range is \"{common_rating}\" with a total count of {common_rating_count}")
```

### Bar Chart :

```python
plt.figure(figsize=(5,5))
plt.hist(rest_data["Aggregate rating"], bins=30, color="skyblue" , edgecolor = "black")
plt.xlabel("Aggregate rating")
plt.ylabel("Counts")
plt.title("distribution of aggregate ratings")
plt.show()
```

### 2. Calculate the average number of votes received by restaurants.


```python
avg_votes = round(rest_data["Votes"].mean(),3)
print(f"The average number of votes recieved by the restaurants is : {avg_votes}")
```
