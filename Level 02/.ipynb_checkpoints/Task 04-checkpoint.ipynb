```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rest_data = pd.read_csv(filepath_or_buffer=r"C:\Users\91882\Desktop\Cognifyz\Dataset .csv",encoding="ISO-8859-1")
```

# Level 02

## Task 04 : Restaurant Chains

### 1. Identify if there are any restaurant chains present in the dataset.

```python
restaurant_counts = rest_data["Restaurant Name"].value_counts()
rest_chain = restaurant_counts[restaurant_counts > 1]
rest_chain_counts = len(rest_chain)
```

```python
print(f"1. There are totally {rest_chain_counts} Restaurant Chains present in the dataset .")
print(f"2. The Restaurant Chain includes : {rest_chain} \n")
```

### Bar Chart :

```python
plt.figure(figsize=(5,5))
top_5_restChain = rest_chain.head(5)
plt.bar(top_5_restChain.index , top_5_restChain.values)
plt.title("Top 5 Restaurant Chains by number of outlets")
plt.xticks(rotation = 90 , ha ="right")
plt.xlabel("Restaurant Name")
plt.ylabel("No. of Chains")
plt.show()
```

### 2. Analyze the ratings and popularity of different restaurant chains

```python
ratings_rest_chains = rest_data.groupby(by="Restaurant Name")["Aggregate rating"].mean().reset_index(name = "Average Rating").sort_values(by = "Average Rating" , ascending=False)
print(f"1. The ratings of different chain restaurants :\n\n{ratings_rest_chains} ")
```

```python
votes_popularity = rest_data.groupby("Restaurant Name")["Votes"].sum().reset_index(name = "Total Votes").sort_values(by = "Total Votes" , ascending=False)
print(f"The popularity of different chain restaurants based on the total votes are : \n\n{votes_popularity}")
```
