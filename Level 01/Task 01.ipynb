## Task 01 : Top Cuisines




### 1. Determine the top three most common cuisines in the dataset.  

```python
cuisine_series = rest_data["Cuisines"].str.split(",").dropna().explode('Cuisines').value_counts()
cuisine_series
```

```python
top_3_Cuisines = cuisine_series.head(3)
print("The top three most common cuisines in the dataset are : ")
for cuisine,count in top_3_Cuisines.items() :
  print(f"{cuisine} : {count}")
```

```python
colors = ["orange" , "blue" , "green"]
plt.bar(top_3_Cuisines.index , top_3_Cuisines.values,color = colors)
plt.title("Top 3 Cuisines")
plt.xlabel("Cuisines")
plt.ylabel("Count")
plt.show()
```

### 2.Calculate the percentage of restaurants that serve each of the top cuisines.

```python
top_5_Cuisines = cuisine_series.head()
total_rest = len(rest_data)
rest_percentage = ((top_5_Cuisines)/total_rest) * 100
print(" percentage of restaurants that serve each of the top cuisines.")
for cuisines , percentage in rest_percentage.items():    
    print(f"{cuisines} : {percentage :.2f}%")
```

```python
colours=["green" , "blue" , "yellow","orange" , "red"]
plt.bar(rest_percentage.index , rest_percentage.values , color=colours)
plt.xlabel("Cuisin Name")
plt.ylabel("Percentage %")
plt.title("Top Serving Cuisines %")
plt.show()
```
