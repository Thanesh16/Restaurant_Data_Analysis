```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rest_data = pd.read_csv(r"C:\Users\91882\Desktop\Cognifyz\Dataset .csv")
```

# Level 01

## Task 04 : Online Delivery

### 1.Determine the percentage of restaurants that offer online delivery .

```python
online_delivery = rest_data["Has Online delivery"].value_counts()
total_rest = len(rest_data['Restaurant Name'])
percet_online_delivery = round(((online_delivery/total_rest)*100),2)
print(f"The percentage of restaurant that offers online delivery is : {percet_online_delivery['Yes']} %")
```

### Pie Chart :

```python
plt.pie(percet_online_delivery , labels = percet_online_delivery.index, autopct="%1.1f%%",colors = ["lightblue" , "lightgreen"])
plt.show()
```

### 2. Compare the average ratings of restaurants with and without online delivery .

```python
avg_ratings = round(rest_data.groupby("Has Online delivery")["Aggregate rating"].mean(),2)
print(f"Average rating with delivery : {avg_ratings['Yes']}")
print(f"Average rating without delivery : {avg_ratings['No']}")
```

### Bar Graph :

```python
plt.figure(figsize=(5,5))
labels = ["With Delivery","Without Delivery"]
plt.bar(labels , avg_ratings.values, color = ["lightgreen" , "lightblue"])
plt.xlabel("Delivery Type")
plt.ylabel("Average ratings")
plt.title("average ratings of restaurants")
plt.show()
```
