```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rest_data = pd.read_csv(filepath_or_buffer = "Dataset .csv")
```

# Level 01

## Task 02 : City Analysis

### 1. Identify the city with the highest number of restaurants in the dataset.

```python
Top_cities= rest_data.groupby(by = "City")["Restaurant Name"].count().sort_values(ascending = False)
high_restaurants = Top_cities.head(1)
print(f"the city with the highest number of restaurants is {high_restaurants} ")
```

```python
top_3 = Top_cities.head(3)
top_3
```

## Bar Chart :

```python
plt.figure(figsize=(7,4))
plt.bar(top_3.index , top_3.values)
plt.text(0 , top_3.values[0] + 50 , top_3.values[0] , ha = "center" , va = "bottom" )
plt.scatter(0 , top_3.values[0] , s=1200 , facecolor = "none" , edgecolors = "black")
plt.ylim(0 , top_3.values[0]+800)
plt.xlabel("Cities")
plt.ylabel("No of Restaurants")
plt.title("Top cities by number of restaurants")
plt.show()
```

### 2. Calculate the average rating for restaurants in each city

```python
avg_rating = rest_data.groupby(by = "City")["Aggregate rating"].mean()
print(f"The average rating of each city is :{avg_rating}")
```

### 3. Determine the city with the highest average rating.

```python
sort_avg_rating = avg_rating.sort_values(ascending=False)
high_avg_city = sort_avg_rating.idxmax()
high_avg_rating = sort_avg_rating.max()
print(f"The city with highest average rating is \"{high_avg_city}\" with a rating value of \"{high_avg_rating}\"")
```

## Bar Chart :

```python
top_3_rating = sort_avg_rating.head(3)
top_3_rating
```

```python
plt.figure(figsize=(7,4))
plt.ylim(0,top_3_rating.values[0] + 1.5)
plt.bar(top_3_rating.index , top_3_rating.values)

for i , value in enumerate (top_3_rating.values):
    plt.text(i,value , value , ha="center" , va = "bottom" )

    
plt.scatter(0 , top_3_rating.values[0] , s=1000 , facecolor ="none" , edgecolor = "black")
plt.xlabel("City")
plt.ylabel("Average rating")
plt.title("highest average rating (cities)")
plt.show()
```
