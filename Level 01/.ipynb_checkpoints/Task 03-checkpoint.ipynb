```python
rest_data = pd.read_csv(filepath_or_buffer = r"C:\Users\91882\Desktop\Cognifyz\Dataset .csv")
```

## Task 03 : Price Range Distribution

### 1. Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.

### Histogram :

```python
plt.figure(figsize=(7,4))
plt.hist(rest_data["Price range"])
plt.title("distribution of price ranges")
plt.xlabel('Price range')
plt.ylabel("Counts")
plt.show()
```

### 2. Calculate the percentage of restaurants in each price range category .

```python
price_range_counts = rest_data["Price range"].value_counts()
price_range_counts
```

```python
total_rest = len(rest_data["Restaurant Name"])
price_range_percentage = round((price_range_counts/total_rest)*100 ,2) 
print(f"The percentage of restaurants in each price range category :{price_range_percentage.astype(str) + '%'}")
```

### Pie Chart

```python
plt.figure(figsize=(5,5))
plt.pie(price_range_percentage, autopct='%1.1f%%' , labels = price_range_percentage.index.astype(float), colors=["yellow" , "blue" , "lightgreen" , "red"])
plt.show()
```
